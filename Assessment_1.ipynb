# Dataset for your Data Science coursework, generate a CSV file named city_weather.csv.


```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Define the date range
start_date = datetime(2023, 7, 31)
end_date = datetime(2024, 7, 30)
date_range = pd.date_range(start_date, end_date)

# Generate synthetic data
np.random.seed(0)  # For reproducibility
temperature = np.random.uniform(low=15, high=35, size=len(date_range))  # Degrees Celsius
humidity = np.random.uniform(low=30, high=90, size=len(date_range))  # Percentage
aqi = np.random.uniform(low=0, high=300, size=len(date_range))  # AQI values

# Create DataFrame
data = {
    'Date': date_range,
    'Temperature': temperature,
    'Humidity': humidity,
    'AQI': aqi
}
df = pd.DataFrame(data)

# Save DataFrame to CSV
df.to_csv('city_weather.csv', index=False)
```

#  Analyzing the city_weather.csv dataset using Python and the libraries NumPy, Pandas, and Matplotlib.

```python
import pandas as pd

# Define the sample data
data = {
    'Date': ['2023-01-01', '2023-01-02', '2023-01-03'],
    'Temperature': [15.2, 16.4, 15.0],
    'Humidity': [45, 50, 47],
    'AQI': [75, 80, 82]
}

# Create DataFrame
df = pd.DataFrame(data)

# Save DataFrame to CSV
df.to_csv('city_weather.csv', index=False)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('city_weather.csv')

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Display basic information about the dataset
print(df.info())
print(df.describe())

# Plot Temperature, Humidity, and AQI over Time
plt.figure(figsize=(14, 7))

# Plot Temperature
plt.subplot(3, 1, 1)
plt.plot(df['Date'], df['Temperature'], marker='o', color='r', linestyle='-')
plt.title('Daily Average Temperature')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.grid(True)

# Plot Humidity
plt.subplot(3, 1, 2)
plt.plot(df['Date'], df['Humidity'], marker='o', color='b', linestyle='-')
plt.title('Daily Average Humidity')
plt.xlabel('Date')
plt.ylabel('Humidity (%)')
plt.grid(True)

# Plot AQI
plt.subplot(3, 1, 3)
plt.plot(df['Date'], df['AQI'], marker='o', color='g', linestyle='-')
plt.title('Daily Average AQI')
plt.xlabel('Date')
plt.ylabel('AQI')
plt.grid(True)

plt.tight_layout()
plt.show()

# Statistical Analysis
print("Correlation Matrix:")
print(df.corr())

# Example Analysis: Check for the hottest and coldest days
hottest_day = df.loc[df['Temperature'].idxmax()]
coldest_day = df.loc[df['Temperature'].idxmin()]

print(f"Hottest Day:\n{hottest_day}")
print(f"Coldest Day:\n{coldest_day}")

# Example Analysis: Average AQI during hottest and coldest days
avg_aqi_hottest_day = df[df['Date'] == hottest_day['Date']]['AQI'].mean()
avg_aqi_coldest_day = df[df['Date'] == coldest_day['Date']]['AQI'].mean()

print(f"Average AQI on Hottest Day: {avg_aqi_hottest_day}")
print(f"Average AQI on Coldest Day: {avg_aqi_coldest_day}")
```

# Calculating average , median , standard deviation of temperature , humidity and AQI in the city over the past year?


```python
import pandas as pd

# Load the dataset
df = pd.read_csv('city_weather.csv')

# Convert 'Date' column to datetime format (if not already)
df['Date'] = pd.to_datetime(df['Date'])

# Calculate statistics
temperature_mean = df['Temperature'].mean()
temperature_median = df['Temperature'].median()
temperature_std = df['Temperature'].std()

humidity_mean = df['Humidity'].mean()
humidity_median = df['Humidity'].median()
humidity_std = df['Humidity'].std()

aqi_mean = df['AQI'].mean()
aqi_median = df['AQI'].median()
aqi_std = df['AQI'].std()

# Print the results
print("Temperature Statistics:")
print(f"Average: {temperature_mean:.2f} °C")
print(f"Median: {temperature_median:.2f} °C")
print(f"Standard Deviation: {temperature_std:.2f} °C")

print("\nHumidity Statistics:")
print(f"Average: {humidity_mean:.2f} %")
print(f"Median: {humidity_median:.2f} %")
print(f"Standard Deviation: {humidity_std:.2f} %")

print("\nAQI Statistics:")
print(f"Average: {aqi_mean:.2f}")
print(f"Median: {aqi_median:.2f}")
print(f"Standard Deviation: {aqi_std:.2f}")
```

# Comparing the 7-day moving average of temperature with the daily temperature values, you'll follow these steps:


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('city_weather.csv')

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Set 'Date' column as index
df.set_index('Date', inplace=True)

# Calculate the 7-day moving average
df['7-Day Moving Average'] = df['Temperature'].rolling(window=7).mean()

# Plot daily temperature and 7-day moving average
plt.figure(figsize=(14, 7))

plt.plot(df.index, df['Temperature'], label='Daily Temperature', color='blue', alpha=0.5)
plt.plot(df.index, df['7-Day Moving Average'], label='7-Day Moving Average', color='red', linewidth=2)

plt.title('Daily Temperature and 7-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Temperature (°C)')
plt.legend()
plt.grid(True)

plt.show()

# Print the first few rows to check the moving average
print(df[['Temperature', '7-Day Moving Average']].head(10))

```

# To determine the relationship between temperature and AQI, and to check if a higher temperature correlates with a higher AQI, follow these steps:

1. Load and Prepare the Data:
* Ensure that the dataset is properly loaded and that the 'Date' column is in datetime format.

2. Calculate the Correlation:
* Use the Pandas .corr() method to compute the correlation coefficient between temperature and AQI.

3. Visualize the Relationship:
* Create a scatter plot to visually inspect the relationship between temperature and AQI.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv('city_weather.csv')

# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Calculate correlation between Temperature and AQI
correlation = df['Temperature'].corr(df['AQI'])
print(f"Correlation between Temperature and AQI: {correlation:.2f}")

# Plot scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df['Temperature'], df['AQI'], alpha=0.5, edgecolors='w', s=100)
plt.title('Temperature vs. AQI')
plt.xlabel('Temperature (°C)')
plt.ylabel('AQI')
plt.grid(True)
plt.show()
```
